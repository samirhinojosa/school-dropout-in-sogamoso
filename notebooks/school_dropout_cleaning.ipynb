### 5.1 Students age

```python
def get_age(date_of_birth):
    """
    Method to get the student age based on date of birth

    Parameters:
    -----------------
        date_of_birth (datetime64): Date of birth.
        
    Returns:
    -----------------
        age (int) : Student age.
    """
    
    today = date.today()
    age = today.year - date_of_birth.year -\
            ((today.month, today.day) <\
             (date_of_birth.month, date_of_birth.day))
    
    
    return age
```

```python
df_students["EDAD"] = df_students["FECHA_NACIMIENTO"].apply(get_age) 
df_students[["FECHA_NACIMIENTO", "EDAD"]].head()
```

### 5.3 Students state

```python
def get_state(df, id_student, year, state, reason):
    """
    Method to determine the school dropout in all students

    Parameters:
    -----------------
        df (pandas.DataFrame): Dataset to analyze.
        id_student (int): student id.
        year (int): year to analyze.
        state (string): Current state.
        reason (string): Reason in case of dropout.
        
    Returns:
    -----------------
        state (boolean) : 1 dropout. 0 not dropout.
    """
    
    # by default not school dropout
    final_state = 0
    
    if year!=2022:
        
        # If the student exist the next year. 
        # 1 - the students exists. 
        # 0 - the students does not exist.
        dropout_following_year = df[(df["ANO"]==year+1) & (df["PER_ID"]==id_student)].shape[0]

        if state=="RETIRADO" and reason=="DESERCION":
            final_state = 1
        elif dropout_following_year == 0:
            final_state = 1
        
    return final_state
```

```python
df_students["ESTADO_FINAL"] = df_students[["PER_ID", 
                                           "ANO",
                                           "ESTADO",
                                           "MOTIVO"]].apply(lambda row: get_state(df_students,
                                                                                  row["PER_ID"],
                                                                                  row["ANO"],
                                                                                  row["ESTADO"],
                                                                                  row["MOTIVO"]),
                                                            axis=1)
```

### 5.4 School size per year

We are going to add a new feature to define the school size considering the number of student by year for each school

```python
df_students_by_school = df_students.groupby(["INSTITUCION", "ANO"])["PER_ID"].count()\
                            .reset_index().rename(columns={"PER_ID" : "ESTUDIANTES"})\
                            .sort_values(by=["ANO", "ESTUDIANTES"], ascending=[True, False])

df_students_by_school
```

```python
# School sizes
group_names = ["Pequeña",  "Mediana", "Grande"]

df_school_bins_size = pd.DataFrame()

for year in range(2013, 2022+1):
    
    # Filtering data based on year
    df_temp = df_students_by_school[df_students_by_school["ANO"]==year].copy()
    
    # Defining the bins/percentiles based on the number of students in the year
    bins = np.linspace(min(df_temp["ESTUDIANTES"]), max(df_temp["ESTUDIANTES"]), 4)
    
    df_temp["INSTITUCION_TAMAÑO"] = pd.cut(df_temp["ESTUDIANTES"], bins, 
                                           labels=group_names, include_lowest=True)
    
    df_school_bins_size = pd.concat([df_school_bins_size, df_temp])
```

```python
df_students = pd.merge(df_students, df_school_bins_size[["INSTITUCION", "ANO", "INSTITUCION_TAMAÑO"]], on=["INSTITUCION", "ANO"], how="left")
```
