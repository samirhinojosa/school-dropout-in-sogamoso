#### 3.1.4 Removing unnecessary columns

Now, let's remove some unnecessary columns identified by Sogamoso's Municipality and also, thanks to initial analysis. 

```python
COLUMNS_TO_DELETE = [
    "DANE", "CONSECUTIVO", "GRUPO", "FECHAINI", "FECHAFIN",  "Unnamed: 25"
]

for col in COLUMNS_TO_DELETE:
    for year in range(2013, 2022+1):
        if col in globals()["df_students_" + str(year)]:
            del globals()["df_students_" + str(year)][col]
```

```python
for year in range(2013, 2022+1): 
    print("df_students_" + str(year))
    print(list(globals()["df_students_" + str(year)].columns),"\n")
```

#### 3.1.5 Completing data between data sets

Now, we are going to complete missing-values between the datasets using the variable `PER_ID` to see if a student have some missing-value that could be fill with data from other dataset.

```python
for year in range(2013, 2022+1):
    percentage_of_NaN = percentage_missing_values(globals()["df_students_" + str(year)])
    print("df_students_" + str(year) + " NaN = " + str(percentage_of_NaN) + " %")
```

Let's see the columns with missing-values for each dataset

```python
for year in range(2013, 2022+1):
    print("df_students_" + str(year))
    print(list(globals()["df_students_" + str(year)].columns[globals()["df_students_" + str(year)].isnull().any()]),"\n")
```

```python
# Completing data 
LIST_OF_DATASETS = []
for year in range(2013, 2022+1):
    LIST_OF_DATASETS.append("df_students_" + str(year))
    
for a, b in itertools.combinations(LIST_OF_DATASETS, 2):
    globals()[a] = globals()[a].set_index("PER_ID").fillna(globals()[b].set_index("PER_ID")).reset_index()
```

```python
for year in range(2013, 2022+1):
    percentage_of_NaN = percentage_missing_values(globals()["df_students_" + str(year)])
    print("df_students_" + str(year) + " NaN = " + str(percentage_of_NaN) + " %")
```
